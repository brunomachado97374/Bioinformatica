### Secção 2 : Redução de Dimensionalidade

#### Técnicas para redução de dimensionalidade:

A redução de dimensionalidade é crucial na aplicação de modelos de Aprendizagem Automática. Em datasets de grandes dimensões, a diminuição do número de features não apenas melhora o desempenho dos modelos, tornando-os mais práticos, mas também reduz a necessidade de recursos computacionais significativos.

Além disso, é vital evitar o overfitting, um fenômeno em que o modelo funciona bem com dados de treino, mas falha com dados inéditos. Ao reduzir a dimensão e complexidade do modelo, concentramo-nos nas features mais informativas, mitigando o risco de overfitting.

A visualização e interpretabilidade do modelo também são beneficiadas pela redução de dimensionalidade. A projeção dos dados em espaços de menor dimensão facilita a identificação de padrões e relações entre as features do modelo e suas previsões.

Assim, a redução de dimensionalidade visa encontrar o menor conjunto de variáveis capazes de descrever a máxima variabilidade dos dados, diminuindo a quantidade de dados a serem avaliados.

#### Normalização dos Dados:

A normalização não é uma técnica de diminuição de dimensionalidade. No entanto, é um processo que ajusta as escalas dos dados, sendo aplicado a características numéricas. Dado que o dataset apresenta-se com diversas variáveis desta categoria, a normalização é um processo extremamente importante.

```python
#Retirar Atributos não tão significativos e atributos que não se encontram no formato float, nomeadamente o Drug_ID, Drug e a nossa label "Y"
from sklearn.preprocessing import MinMaxScaler

pd.set_option('display.max_row', 10)

atributos=d.iloc[:, 2:-1]
atributos


scaler = MinMaxScaler(feature_range=(-1, 1))
atributos= pd.DataFrame(scaler.fit_transform(atributos), columns=atributos.columns)
atributos
```

##### VarianceThreshold: 

O objetivo passa por identificar as features que têm pouca ou nenhuma variação, pois não contribuem significativamente para a modelação. Se uma feature tem baixa variação, significa que os seus valores são relativamente constantes em todo o conjunto de dados, o que indica que ela não fornece muita informação útil para distinguir entre diferentes amostras.

```python
threshold = 0.01

from sklearn.feature_selection import VarianceThreshold

var_threshold = VarianceThreshold(threshold=threshold)
x_variance = var_threshold.fit_transform(atributos)

variance_columns = atributos.columns[var_threshold.get_support()]


high_variance_data = atributos[variance_columns]
print("Features com maior variância:\n", high_variance_data.shape)
high_variance_data.head()
```

#### Seleção de dados por percentil:

Refere-se a um método no qual os valores de uma variável são escolhidos com base na sua posição percentual em relação ao conjunto total de dados.

Esta abordagem é útil em várias situações,nomeadamente, a remoção de valores extremos (outliers) ou a concentração numa faixa específica de valores.

```python
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectPercentile, f_regression

selector = SelectPercentile(f_regression, percentile=50)

selector.fit(atributos, d["Y"])

selected_features = selector.transform(atributos)

df_percent = pd.DataFrame(selected_features, columns=atributos.columns[selector.get_support()])

df_percent
```

#### SelectKBest:

O SelectKBest é uma estratégia de filtragem que escolhe os k melhores recursos (ou colunas) de um conjunto de dados, visando reter apenas os mais relevantes. Esta técnica avalia cada recurso individualmente mediante uma métrica escolhida (F-value) e, em seguida, seleciona os k recursos com a pontuação mais elevada.

```python
from sklearn.feature_selection import SelectKBest, f_classif
import pandas as pd

labels = d['Y']
features = d.drop(['Y', 'Drug_ID', 'Drug'], axis=1)  

skb = SelectKBest(f_classif, k=10)
x_kbest = skb.fit_transform(features, labels)

selected_indices = skb.get_support(indices=True)

selected_columns = features.columns[selected_indices]

data_kbest = pd.DataFrame(data=x_kbest, index=features.index, columns=selected_columns)
```

```python
import matplotlib.pyplot as plt

# Criar o boxplot dos dados selecionados com os 10 melhores amostras
plt.boxplot(data_kbest)
plt.title('10 melhores amostras')
plt.xticks(np.arange(1, 11), data_kbest.columns, rotation=90)
plt.show()
```

#### PCA - Principal Component Analysis:

A Análise de Componente Principais (PCA) é uma técnica de redução de dimensionalidade, tendo como objetivo  a transformação de um dataset extenso e complexo num novo sistema de coordenadas (componentes principais), atendendo a que a variância dos dados é maximizada. O PCA assume relações lineares entre features, o que lhe confere também a capacidade de estabelecer combinações de features através destas relações, de tal modo a representar os dados de forma mais informativa.

Assim, a aplicação da técnica de PCA permite captar as informações (features) de maior relevância à medida que reduz a dimensão dos dados. Com isto, consegue-se uma representação de dados computacionalmente mais eficientes e proporciona-se uma visualização facilitada dos dados de elevada dimensão para uma melhor análise e interpretação.

```python
from sklearn.decomposition import PCA


n = 20
pca = PCA(n_components=n)
pca.fit(atributos)
x_pca = pca.transform(atributos)
x_pca=pd.DataFrame(x_pca)
x_pca
```

```python
plt.figure(figsize=(10,5))
plt.bar(range(n), pca.explained_variance_ratio_*100)
plt.xticks(range(n), ['PC'+str(i) for i in range(1,n+1)])
plt.title("Variância Representada por PC")
plt.ylabel("Percentagem")

plt.show()
```

Através do gráfico demonstrando acima, resultante da análise PCA, entende-se que as features apresentam baixa capacidade de representar a variância dos dados, uma vez que tem-se um máximo de representação de variância pouco acima dos 25%, o que indica a necessidade de avaliar um conjunto mais extenso de PCs para representar corretamente os dados do caso.

### Secção 3 : Modelos de Aprendizagem 

#### Aprendizagem Supervisionada

A abordagem da aprendizagem supervisionada no contexto do nosso estudo envolve treinar modelos utilizando um conjunto de dados em que as entradas estão associadas às saídas correspondentes. No nosso caso, o modelo é treinado com base nas informações das fórmulas químicas de fármacos, procurando perceber a sua relação com a penetração da BBB. O conjunto de dados inclui características dos fármacos e a variável target, que indica se um fármaco penetrou a barreira (1) ou não (0). Este é um problema de classificação supervisionada, pois estamos interessados em prever a resposta ao tratamento com base nos dados de expressão genética.

Durante o treinamento do modelo, o algoritmo utiliza exemplos passados, onde a resposta ao tratamento é conhecida, para aprender a relação entre as características e o sucesso na penetração da barreira. Posteriormente, o modelo é capaz de generalizar esse conhecimento para prever a resposta para novos fármacos com base nas suas características. Portanto, neste contexto, iremos recorrer a modelos de aprendizagem supervisionada, dado que a variável target é discreta, indicando a sua natureza binária.

Para a realização deste projeto, iremos realizar dois processos que serão dividos utilizando dois métodos diferentes (Split Data e Cross-Validation) e, posteriormente, aplicados a 5 modelos diferentes, sendo eles, Logistic Regression, Decision Tree, Random Forest, Naive Bayes e SVC-Support Vector Classification.

- Processo 1 : Dataset inicial normalizado;
- Processo 2 : Dataset proveniente do PCA;

#### Aplicação de Modelos Supervisionados ao Processo 1:

Para posterior avaliação dos modelos, que consiste num processo que mede o desempenho e eficácia dos algoritmos aplicados a dados de teste, ou seja, a dados não vistos durante o processo de treino, serão calculadas para todos os modelos aplicados a ambos os processos diversas métricas de avaliação. 

Matriz de Confusão: Tabela que compara as previsões do modelo com os valores reais do conjunto de teste, apresentando o número de verdadeiros positivos (TP), verdadeiros negativos (TN), falsos positivos (FP) e falsos negativos (FN). A partir desta é possível determinar:

  - Accuracy: Proporção de previsões corretas em relação ao total de previões;  

  - Precision: Razão entre os verdadeiros positivos e o total de instâncias previstas como positivas;  

  - Recall(Sensibilidade): Proporção entre os verdadeiros positivos e o total de instâncias positivas no conjunto de dados.  

  - F1-Score: Combina precisão e recall num único valor de avaliação dos modelos. É utilizada em problemas de classificação desbalanceados, onde uma classe apresenta mais prevalência do que a outra e é particularmente útil em situações em que tanto falsos positivos quanto falsos negativos são importantes.

#### Aplicação de Modelos Supervisionados ao Processo 1, com a particição de dados através do Cross-Validation:

O cross-validation, é uma técnica utilizada em machine learning para avaliar o desempenho de um modelo. O objetivo é maximizar a utilização dos dados disponíveis, principalmente em casos em que o conjunto de dados é limitado.

Esta envolve a divisão do conjunto de dados em partes, chamadas "folds". O modelo é treinado em várias iterações, onde em cada iteração, um dos folds é utilizado como conjunto de teste, enquanto os folds restantes são usados como conjunto de treino. Esse processo é repetido até que todos os folds tenham sido utilizados como conjunto de teste. Para os processos utilizou-se um número de folds k=5. A validação cruzada é crucial para estimar a performance de um modelo de forma mais confiável, especialmente quando o conjunto de dados é limitado, evitando problemas como overfitting ou subestimação do desempenho do modelo.

#### Ensemble Model:

Ideia central da construção de ensemble models reside na obtenção de respostas a situações novas, a partir da combinação das respostas de diferentes modelos. De facto, a construção de ensemble models permite combinar as previsões de vários modelos individuais de modo a obter um desempenho otimizado, relativamente do que o desempenho individual. Assim, usando os diferentes modelos desenvolvidos posibilita o aprimoramento da precisão e da robustez dos resultados.

#### Voting Classifier:

A utilização do Voting Classifier, na construção de um ensemble model, permite combinar vários modelos de aprendizagem supervisionada para tomar decisões por votação. Nos modelos desenvolvidos foi considerada a "hard voting", na medida em que as previsões de cada modelo são consideradas no ensemble model, e a classe que recebe a maioria dos votos, dos modelos individuais, é selecionada como a previsão final.

#### Aplicação do Voting Classifier ao Processo 1 com partição de Dados Split Data:

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import VotingClassifier
import seaborn as sns
import matplotlib.pyplot as plt
from prettytable import PrettyTable

def create_voting_classifier(models):
    voting_models = [(name, model) for name, model in models]
    voting_clf = VotingClassifier(estimators=voting_models, voting='soft')
    return voting_clf

model_list = [
    ('SVC', SVC_norm),
    ('RandomForest', RFC_norm),
    ('NaiveBayes', NB_norm),
]

voting_classifier = create_voting_classifier(model_list)

voting_classifier.fit(X_train_norm, y_train_norm)
previsoes_voting = voting_classifier.predict(X_test_norm)

relatorio_voting = classification_report(y_test_norm, previsoes_voting, output_dict=True)


linha_relatorio_Ensemble = {
    'Modelo': 'Voting Classifier',
    'Divisão de Dados': 'SD',
    'Accuracy': str(relatorio_voting['accuracy']),
    'Precision': str(relatorio_voting['1']['precision']),
    'Recall': str(relatorio_voting['1']['recall']),
    'F1-Score': str(relatorio_voting['1']['f1-score'])
}

tabela_ensemble = PrettyTable()
tabela_ensemble.field_names = linha_relatorio_Ensemble.keys()
tabela_ensemble.add_row(linha_relatorio_Ensemble.values())

print("\nTabela do Relatório:")
print(tabela_ensemble)

matrix_confusao_voting = confusion_matrix(y_test_norm, previsoes_voting)

plt.figure(figsize=(2, 2))
sns.heatmap(matrix_confusao_voting, annot=True, fmt='d', cmap='Blues', cbar=False, linewidths=0.5)
plt.title('Matriz de Confusão - Voting Classifier')
plt.xlabel('Previsto')
plt.ylabel('Verdadeiro')
plt.show()
```

O código anterior permite a criação de um ensemble model, combinando 3 modelos diferentes.

Comparando o desempenho do ensemble model com os modelos individuais constata-se que este obteve uma accuracy média de 84%, valor próximo ao dos modelos inividuais mais fortes. Devido ao modelo de Naive Bayes apresentar uma menor accuracy média, a sua incorporação no ensemble model poderá ter contribuído para piorar a robustez geral.
