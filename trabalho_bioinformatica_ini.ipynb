```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from tdc.single_pred import ADME

# Carregar o conjunto de dados
data = ADME(name='BBB_Martins')
df = data.get_data().head(1976)

# Análise exploratória inicial
print("Informações sobre o conjunto de dados:")
print(df.info())

# Descrição estatística das variáveis numéricas
print("\nDescrição estatística das variáveis numéricas:")
print(df.describe())

# Verificar dados ausentes
print("\nValores ausentes no conjunto de dados:")
print(df.isnull().sum())

# Análise exploratória com alguns gráficos
sns.set(style="whitegrid")

# Gráfico de contagem para a variável alvo (Y)
plt.figure(figsize=(6, 4))
sns.countplot(x='Y', data=df)
plt.title('Distribuição da Variável Alvo (Y)')
plt.show()

# Histograma do número de caracteres nas fórmulas químicas (Drug)
df['Drug_Length'] = df['Drug'].apply(len)

plt.figure(figsize=(10, 6))
sns.histplot(df['Drug_Length'], bins=20, kde=False)
plt.title('Distribuição do Número de Caracteres nas Fórmulas Químicas (Drug)')
plt.xlabel('Número de Caracteres')
plt.ylabel('Contagem')
plt.show()

# Gráfico de densidade
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Drug_Length'], color='red', fill=True)
plt.title('Distribuição do Número de Caracteres nas Fórmulas Químicas (Drug)')
plt.xlabel('Número de Caracteres')
plt.show()

# Remover a coluna adicionada para evitar interferência em etapas posteriores
df.drop('Drug_Length', axis=1, inplace=True)



```
